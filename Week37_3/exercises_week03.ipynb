To add a legend (i.e., naming the classes) we sort the data into the two classes before plotting. We may then use the `plot` command.

````python
def show(X, y, marker='.'):
    labels = set(y)
    cl = {lab : [] for lab in labels}
    # cl[lab] shall contain the datapoints labeled lab
    for (a, b) in zip(X, y):
        cl[b].append(a)
    for lab in labels:
        plt.plot([a[0] for a in cl[lab]], [a[1] for a in cl[lab]], 
                 marker, label="class {}".format(lab))
    plt.legend()
````

### Exercise I: Distance
Implement a (L2-) distance function. It should work for points in *n*-dimensional space for any integer *n*>0. Check that dist((3,4,0),(0,0,12)) is 13.

````python
def distance_L2(a,b):
    """
    Calculate and return the L2-distance between a and b, both being n dimensional tuples"
    """
    if len(a) != len(b):
        raise ValueError("Error: a and b must be of the same dimension")
    s = 0
    for i in range(len(a)):
        s += (b[i] - a[i])**2
    return np.sqrt(s)
````

````python
assert distance_L2((3,4,0),(0,0,12)) == 13
````

### Exercise II: Majority class
The next thing we need is a way to determine the majority class from a set of votes. Implement a procedure which takes a list as argument and returns the majority class. 

````python
def majority(a):
    """
    Return the majority class of a, i.e. the most common item in tuple a.
    For example majority([0,1,1,1,0]) should return 1
    """
    from collections import Counter
    counts = Counter(a)
    return counts.most_common(1)[0][0]
````

````python
assert majority([0,1,1,1,0]) == 1
````

````python
class PykNNClassifier(PyClassifier):
    """kNN classifier using pure python representations"""
    
    def __init__(self, k=3, dist=distance_L2):
        self.k = k
        self.dist = dist
        
    def fit(self, X_train, t_train):
        self.X_train = X_train
        self.t_train = t_train
    
    def predict(self, a):
        """Implement this"""
        if type(a) != tuple:
            print("a must be a 1D tuple with 2 elements (x, y)")
        else:
            distance = []
            # Fill d with tuples containing: (index in X_train, and distance from a)
            #for i in range(len(self.X_train)):
            #    xy = self.X_train[i]    # tuple representing the point (x, y) of index i in X_train
            #    distance_a_xy = self.dist(a, xy)
            #    distance.append((i, distance_a_xy)) # add tuple (i, dist) where i refers to the point in X_train with index i 
            for i, x in enumerate(self.X_train): # enumerate put an index i for each element in X_train    
                d = self.dist(a, x)
                distance.append((i, d)) # store tuple (i, d) in distance, where i refers to the point in X_train with index i
            print(distance)
            # Sort the distance elements by distance
            # Since we keep track of the index, we can use .sort()
            def get_distance(tup):
                """
                Extract the second element, from a tuple x.
                """
                return tup[1]
            distance.sort(key=get_distance) # sorts in ascending order
            print(distance)
            # Store k first elements of distance to a k_nearest_list
            k_nearest_list = distance[:self.k]
            
            # Finding the t_values with corresponding index to the idex element of k_nearest_list elements
            t_values_k_nearest = []
            for i in range(len(k_nearest_list)):
                t_values_k_nearest.append(self.t_train[k_nearest_list[i][0]])
            print(t_values_k_nearest)
            # Evaluate and return most common value, i.e. class
            return majority(t_values_k_nearest)
````
