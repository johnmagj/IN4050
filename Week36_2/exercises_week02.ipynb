### 1. Given the function $ f(x) = -x^{4} + 2x^{3} + 2x^{2} - x $ and it's derivative $f'(x) = {- 4 x^3 + 6 x^2 + 4 x - 1}$

```python
# Standard import and functions
# Run this cell first
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return -np.power(x, 4) + 2 * np.power(x, 3) + 2 * np.power(x, 2) - x

def df(x):
    return -4 * np.power(x, 3) + 6 * np.power(x, 2) + 4 * x - 1
```

#### a. Plot the function, and its gradient (derivative) from $x = -2$ to $x = 3$

```python
# Solution:

x = np.linspace(-2, 3, 1000)
plt.plot(x, f(x), label="f(x)")
plt.plot(x, df(x), label="df(x)")
plt.xlabel("x")
plt.ylabel
plt.legend()
plt.grid()
plt.show()
plt.close()
```

#### b. Maximize using gradient ascent. You can try step size 0.1 and start somewhere in the range $[-2, 3]$. 

Gradient ascent: $x_{(k+1)} = x_{(k)} + \gamma \nabla f(x_{(k)})$

```python
# Solution

def gradient_ascent(x_start, f, df, precision, learning_rate):
    """
    f and df must both be functions of x, i.e. f(x), df(x), where df=df/dx
    Calculate x_current, using the learning rate lr. Terminate when set precision is met.
    """
    gamma = learning_rate
    df_current = df(x_start)
    print(f"Start value of dfdx = {df_current}")
    x_current = x_start
    i = 0
    while abs(df_current) > precision:
        x_current = x_start + gamma*df(x_start)
        x_start = x_current
        df_current = df(x_current)
        i += 1
    print(f"Final df/dx = {df_current}")
    print(f"Number of iterations: {i}")
    return x_current, f(x_current)

lr = 0.1
precision = 0.0001
x_start = 2.9

found_x, found_f = gradient_ascent(x_start, f, df, precision=precision, learning_rate=lr)

plt.scatter(x_start, f(x_start), color="green", marker="x", s=200, label=f"Starting point, x={x_start}")
plt.scatter(found_x, found_f, color="red", marker="*", s=200, label="Found maxima")
plt.plot(x, f(x), label="f(x)", color="#1791e8")
#plt.plot(x, df(x), label="df(x)")
plt.xlabel("x")
plt.ylabel
plt.legend()
plt.title(f"Gradient Ascent with Learning Rate = {lr}, and precision = {precision}")
plt.grid()
plt.show()
plt.close()
```

##### How does the choice of starting point and step size affect the algorithm's performance?

##### Is there a starting point where the algorithm would not even be able to find a local maximum?

**Answer:** First and foremost we see that if the learning rate is set too high ($0.1$ in this case), we overshoot and miss the global maximum even if this is the closest maximum. This happens even if the precision is set quite high, possibly because the global maximum is sharp, and the area where the slope is less than the tolerance is narrow. In the other cases the precision is low ($0.01$), and we can get stuck if we start close to a local minimum or saddle point. Even with low learning rate and high precision, there are staring points that will make this model only able to find the local maximums, like left of the saddle point, and left of the local maxima.
