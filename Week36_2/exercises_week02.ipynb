### 1. Given the function $ f(x) = -x^{4} + 2x^{3} + 2x^{2} - x $ and it's derivative $f'(x) = {- 4 x^3 + 6 x^2 + 4 x - 1}$

```python
# Standard import and functions
# Run this cell first
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return -np.power(x, 4) + 2 * np.power(x, 3) + 2 * np.power(x, 2) - x

def df(x):
    return -4 * np.power(x, 3) + 6 * np.power(x, 2) + 4 * x - 1
```

#### a. Plot the function, and its gradient (derivative) from $x = -2$ to $x = 3$

```python
# Solution:

x = np.linspace(-2, 3, 1000)
plt.plot(x, f(x), label="f(x)")
plt.plot(x, df(x), label="df(x)")
plt.xlabel("x")
plt.ylabel
plt.legend()
plt.grid()
plt.show()
plt.close()
```

#### b. Maximize using gradient ascent. You can try step size 0.1 and start somewhere in the range $[-2, 3]$. 

Gradient ascent: $x_{(k+1)} = x_{(k)} + \gamma \nabla f(x_{(k)})$

```python
# Solution

def gradient_ascent(x_start, f, df, precision, learning_rate):
    """
    f and df must both be functions of x, i.e. f(x), df(x).
    Calculate x_next, using the learning rate lr. Terminate when set precision is met.
    """
    gamma = learning_rate
    dfx = df(x_start)
    x_next = x_start
    i = 0
    while abs(dfx) > precision:
        x_next = x_start + gamma*df(x_start)
        x_start = x_next
        dfx = df(x_next)
        i += 1
    print(dfx)
    print(f"Number of iterations: {i}")
    return x_next, f(x_next)

lr = 0.1
precision = 0.0001
x_start = 0.25

testx, testf = gradient_ascent(x_start, f, df, precision=precision, learning_rate=lr)

plt.scatter(x_start, f(x_start), color="red", marker=".", s=200, label=f"Starting point, x={x_start}")
plt.scatter(testx, testf, color="red", marker="*", s=200, label="Found maxima")
plt.plot(x, f(x), label="f(x)", color="#1791e8")
#plt.plot(x, df(x), label="df(x)")
plt.xlabel("x")
plt.ylabel
plt.legend()
plt.title(f"Gradient Ascent with Learning Rate = {lr}, and ")
plt.grid()
plt.show()
plt.close()
```
