### 1. Given the function $ f(x) = -x^{4} + 2x^{3} + 2x^{2} - x $ and it's derivative $f'(x) = {- 4 x^3 + 6 x^2 + 4 x - 1}$

```python
# Standard import and functions
# Run this cell first
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return -np.power(x, 4) + 2 * np.power(x, 3) + 2 * np.power(x, 2) - x

def df(x):
    return -4 * np.power(x, 3) + 6 * np.power(x, 2) + 4 * x - 1
```

#### a. Plot the function, and its gradient (derivative) from $x = -2$ to $x = 3$

```python
# Solution:

x = np.linspace(-2, 3, 1000)
plt.plot(x, f(x), label="f(x)")
plt.plot(x, df(x), label="df(x)")
plt.xlabel("x")
plt.ylabel
plt.legend()
plt.grid()
plt.show()
plt.close()
```

#### b. Maximize using gradient ascent. You can try step size 0.1 and start somewhere in the range $[-2, 3]$. 

Gradient ascent: $x_{(k+1)} = x_{(k)} + \gamma \nabla f(x_{(k)})$

```python
# Solution

def gradient_ascent(x_start, precision=0.0001, lr=0.1):
    """
    Calculate x_next, using the learning rate lr.
    """
    gamma = lr
    fx = f(x_start)
    dfx = df(x_start)
    while abs(dfx) > precision:
        x_next = x_start + gamma*df(x_start)
        fx = f(x_next)
        dfx = df(x_start)
        x_start = x_next
    return x_next, fx

x_start = 0.0
testx, testf = gradient_ascent(x_start)

plt.scatter(testx, testf)
plt.scatter(x_start, f(x_start))
plt.plot(x, f(x), label="f(x)")
plt.plot(x, df(x), label="df(x)")
plt.xlabel("x")
plt.ylabel
plt.legend()
plt.grid()
plt.show()
plt.close()
```
